Chargement du dataset


```python
import pandas as pd
import os

RAW_PATH = "../data/raw_data_cars.csv"
PROCESSED_PATH = "../data/processed_data.csv"
df = pd.read_csv(RAW_PATH)
print("Colonnes détectées :", df.columns.tolist())
print(df.head())
```

Sélection des colonnes importantes

```python

cols_to_keep = ['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 
                'wheelbase', 'curbweight', 'enginetype', 'cylindernumber', 
                'enginesize', 'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price']

df_filtered = df[cols_to_keep]
```

Encodage

```python

df_final = pd.get_dummies(df_filtered, drop_first=True)

print(" Nouvelle forme du dataset :", df_final.shape)
```

Sauvegarde du fichier propre

```python

os.makedirs("../data", exist_ok=True)

df_final.to_csv(PROCESSED_PATH, index=False)
print(f" Fichier sauvegardé sous : {PROCESSED_PATH}")
```
